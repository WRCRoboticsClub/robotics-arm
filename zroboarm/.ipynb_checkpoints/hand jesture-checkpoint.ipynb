```python
import cvzone

from cvzone.HandTrackingModule import HandDetector
import cv2 as cv
```

```python
cap=cv.VideoCapture(0)
detector=HandDetector(detectionCon=0.8,maxHands=1)
while True:
    sucess,img=cap.read()
    hands,img = detector.findHands(img)
    if hands:
        hands1=hands[0]
        lmlist1=hands1['lmList']
        bbox1=hands1["bbox"]
        if lmlist1:
            fingers=detector.fingersUp(hands1)
            print(fingers)
            #fingers detection
            if fingers==[0,1,1,1,1] or fingers==  [1,1,1,1,1]:
                print('expand',fingers)
            if fingers==[0,0,0,0,0] or fingers== [1,0,0,0,0]:
                print('contract',fingers)
            if fingers==[1,1,0,0,0] or fingers==[0,1,0,0,0]:
                print('increase height')
            if fingers==[1,1,1,0,0] or fingers==[0,1,1,0,0]:
                print('decrease height')
    cv.imshow("hand detection",img)
    key=cv.waitKey(1)
    if key==ord('q'):
        cap.release()
        break
cv.destroyAllWindows()
```
